```python
#dependencies
import pandas as pd
import os
import numpy as np
```

```python
#determine whether we need to merge two or more files

passed_merge_input = False
passed_confirmation_input = False
while passed_merge_input != True:
    merge_choice = input("Would you like to use (1) file or merge (2)?")
    if merge_choice == '2':
        passed_merge_input = True
        print("Great let's merge your files.")
    elif merge_choice == '1':
        passed_merge_input = True
    else:
        print("Sorry, I don't understand. Please select a valid choice.")
```

```python
#process fork for when only 1 input file
if merge_choice == '1':
    file_name = input("Let's Analyze your file. Please enter a json file name in the same directory as this program excluding the extension.")                  
    #define usable file_path with user input
    
    file_path = (str(file_name) + ".json")
    
    #read json file and convert into dataframe
    
    full_table_df = pd.read_json(file_path)
    full_table_df
```

```python
#process for 2 input files

while passed_confirmation_input != True:
    if merge_choice == '2':
        file_name = input("File 1 - Please enter a json file name in the same directory as this program, excluding the extension.")
        file_path = (str(file_name) + ".json")
        file_name2 = input("File 2 - Please enter a json file name in the same directory as this program, excluding the extension.")
        file_path2 = (str(file_name2) + ".json")
    confirmed = input("The files you selected are " + str(file_name) + ".json & " + str(file_name2) + ".json. Is this correct? (Y)")
    confirmed = confirmed.lower()
    if confirmed == 'y':
        print("Preparing Output")
        passed_confirmation_input = True
    
full_table_df2 = pd.read_json(file_path)
full_table_df3 = pd.read_json(file_path2)

full_table_df = pd.merge(full_table_df2, full_table_df3, how='inner')
```

```python
full_table_df
```

```python
nPlayers = full_table_df[['SN']].nunique()
TotalPlayers_df = pd.DataFrame({'Total Players':nPlayers})
```

```python
TotalPlayers_df
```

```python
nUniqueItems = full_table_df['Item ID'].nunique()
avgPrice = full_table_df['Price'].mean()
nPurchases = full_table_df.count()
totalRevenue = full_table_df['Price'].sum()

purchasingAnalysis_df = pd.DataFrame({'Unique Items': nUniqueItems, 
                                     'Average Purchase Price': avgPrice,
                                     'Total Number of Purchases': nPurchases,
                                     'Total Revenue':totalRevenue})

purchasingAnalysis_df.head(1)
```

```python
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

gender_df = full_table_df.groupby('Gender')

genderpurchase = gender_df['SN'].count()
genderpurchase_df = pd.DataFrame({
    "Count" : genderpurchase,
    "Percentage" : genderpurchase/779})

genderpurchase_df
```

```python
'''
The below each broken by gender
Purchase Count
Average Purchase Price
Total Purchase Value
Normalized Totals
'''

gender_sum = full_table_df.groupby(by='Gender')['Price'].sum()
gender_avg = full_table_df.groupby(by='Gender')['Price'].mean()
gender_total = full_table_df.groupby(by='Gender')['SN'].count()
gender_total_normal = full_table_df.groupby(by='Gender')['Price'].sum()

gender_analysis_df = pd.DataFrame({
    "Purchase Sum" : gender_sum,
    "Purchase Average" : round(gender_avg, 2),
    "Purchase Count" : gender_total,
    "Normalized Total" : gender_total_normal})

gender_analysis_df = gender_analysis_df[['Purchase Sum', 
                                        'Purchase Average', 
                                        'Purchase Count', 
                                        'Normalized Total']]
gender_analysis_df
```

```python
'''
The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
Purchase Count
Average Purchase Price
Total Purchase Value
Normalized Totals
'''

bins = [0, 9, 14, 19, 24, 29, 34, 39, 44, 49]
full_table_df['Age Bracket'] = pd.cut(full_table_df['Age'], bins)

binned_sum = full_table_df.groupby(by='Age Bracket')['Price'].sum()
binned_avg = full_table_df.groupby(by='Age Bracket')['Price'].mean()
binned_total = full_table_df.groupby(by='Age Bracket')['SN'].count()
binned_total_normal = full_table_df.groupby(by='Age Bracket')['Price'].sum()

binned_analysis_df = pd.DataFrame({
    "Purchase Sum" : binned_sum,
    "Purchase Average" : round(binned_avg, 2),
    "Purchase Count" : binned_total,
    "Normalized Total" : binned_total_normal})

binned_analysis_df = binned_analysis_df[['Purchase Sum', 
                                        'Purchase Average', 
                                        'Purchase Count', 
                                        'Normalized Total']]
binned_analysis_df
```

```python
'''
Top Spenders

Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
SN
Purchase Count
Average Purchase Price
Total Purchase Value
'''

player_num = full_table_df.groupby(by='SN')['SN'].max()
player_count = full_table_df.groupby(by='SN')['Price'].count()
player_avg = full_table_df.groupby(by='SN')['Price'].mean()
player_total = full_table_df.groupby(by='SN')['Price'].sum()

player_analysis_df = pd.DataFrame({
    "Serial Number" : player_num,
    "Purchase Count" : player_count,
    "Average Purchase Price" : player_avg,
    "Total Purchase Value" : player_total})
    
player_analysis_df = player_analysis_df[['Serial Number', 
                                        'Average Purchase Price', 
                                        'Purchase Count', 
                                        'Total Purchase Value']]

player_analysis_df.sort_values(by='Total Purchase Value', ascending=False).head(5)
```

```python
'''
Most Popular Items

Identify the 5 most popular items by purchase count, then list (in a table):
Item ID
Item Name
Purchase Count
Item Price
Total Purchase Value
'''
item_id = full_table_df.groupby(by='Item ID')['Item ID'].max()
item_name = full_table_df.groupby(by='Item ID')['Item Name'].max()
item_count = full_table_df.groupby(by='Item ID')['Price'].count()
item_price = full_table_df.groupby(by='Item ID')['Price'].max()
item_total = full_table_df.groupby(by='Item ID')['Price'].sum()

item_analysis_df = pd.DataFrame({
    "Item ID" : item_id,
    "Item Name" : item_name,
    "Purchase Count" : item_count,
    "Item Price" : item_price,
    "Total Purchase Value" : item_total})
    
item_analysis_df = item_analysis_df[['Item ID', 
                                        'Item Name', 
                                        'Purchase Count', 
                                        'Item Price',
                                        'Total Purchase Value']]

item_analysis_df.sort_values(by='Purchase Count', ascending=False).head(5)
```

```python
'''
Identify the 5 most profitable items by total purchase value, then list (in a table):
Item ID
Item Name
Purchase Count
Item Price
Total Purchase Value
'''
profit_id = full_table_df.groupby(by='Item ID')['Item ID'].max()
profit_name = full_table_df.groupby(by='Item ID')['Item Name'].max()
profit_count = full_table_df.groupby(by='Item ID')['Price'].count()
profit_price = full_table_df.groupby(by='Item ID')['Price'].max()
profit_total = full_table_df.groupby(by='Item ID')['Price'].sum()

profit_analysis_df = pd.DataFrame({
    "Item ID" : profit_id,
    "Item Name" : profit_name,
    "Purchase Count" : profit_count,
    "Item Price" : profit_price,
    "Total Purchase Value" : profit_total})
    
profit_analysis_df = item_analysis_df[['Item ID', 
                                        'Item Name', 
                                        'Purchase Count', 
                                        'Item Price',
                                        'Total Purchase Value']]

profit_analysis_df.sort_values(by='Item Price', ascending=False).head(5)
```
